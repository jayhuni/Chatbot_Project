- 취합데이터 토큰화 / 어간 추출

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import WordPunctTokenizer
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/세훈재훈민준_결측수정.csv')
df.tail(3)
```

```python
# 결측치 체크/ unnamed 제거
df.drop(columns=['Unnamed: 0'], inplace=True)
df.isna().sum()
```

```python
from konlpy.tag import Okt
okt = Okt()
# '질문' 열의 각 행에 대해 명사 추출
nouns_list = df['질문'].apply(lambda x: okt.nouns(x))

# 결과 확인
for index, nouns in enumerate(nouns_list):
    print(nouns)
```

```python
concat_nouns = ' '.join([' '.join(nouns) for nouns in nouns_list])
okt.morphs(concat_nouns, stem=True)
```

- 불용어 제거

```python
with open('data/한글불용어100.txt') as st:
    lines = st.readlines()
print(lines[0])
```

```python
stop_words = []
for line in lines:
    word = line.split('\t')[0]
    stop_words.append(word)
```

```python
# stop_words 추가
additional_stop_words = ['요', '햇', '던데', '식이', '살수', '츄', '꼭알', '소', '꼭', '아야', '여', '를', '듭', '니', '채', '실', '걸', '색', '세', '박', '의', '시', '벌', '과', '라며', '식']
stop_words.extend(additional_stop_words)
```

```python
# 형태소 분석기를 돌려서 나온 결과에서 불용어 제거하기
morphs = okt.morphs(concat_nouns, stem=True)
clean_morphs = [morph for morph in morphs if morph not in stop_words]
clean_morphs
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 한글 문장 임베딩에 활용하기 위해 만들어진 한글 사전학습 모델
model = SentenceTransformer('jhgan/ko-sroberta-multitask')
```

```python
df['embedding'] = pd.Series([[]] * len(df))     # dummy
df.shape
```

```python
df.embedding = df.질문.map(lambda x: list(model.encode(x)))
df.head(3)
```

```python
df.to_csv('data/test_dataset.csv', index=False)
```

```python
import json
wdf = pd.read_csv('data/test_dataset.csv')
wdf.embedding = wdf.embedding.apply(json.loads)
wdf.head(3)
```

```python
text = '매매 팁'
embedding = model.encode(text)
wdf['유사도'] = wdf.embedding.map(lambda x: cosine_similarity([embedding],[x]).squeeze())
answer = wdf.답변[wdf.유사도.idxmax()]
answer
```
